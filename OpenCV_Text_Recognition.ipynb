```python
import cv2
import pytesseract
```

```python

imagem = cv2.imread('/content/Exemplo-melhor.png')

from google.colab.patches import cv2_imshow

cv2_imshow(imagem)


newImage = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
cv2_imshow(newImage)

config_tesseract = "--tessdata-dir tessdata --psm 11"
Texto = pytesseract.image_to_string(imagem, lang= 'por', config= config_tesseract)
print(Texto)
```

```python
!tesseract --help-psm
```

```python
limiar = 127
valor, lim_simples = cv2.threshold(newImage, limiar, 255, cv2.THRESH_BINARY)
cv2_imshow(lim_simples)

config_tesseract = "--tessdata-dir tessdata --psm 11"

Texto = pytesseract.image_to_string(lim_simples, lang= 'por', config= config_tesseract)
print(Texto)
```
